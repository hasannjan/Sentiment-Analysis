```python
from rake_nltk import Rake
import numpy as np
import pandas as pd
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
r = Rake() # Uses stopwords for english from NLTK, and all puntuation characters.

r.extract_keywords_from_text("this movie was worth watching")

r.get_ranked_phrases() # To get keyword phrases ranked highest to lowest.
```

```python
dataset = pd.read_csv('Amazon_Reviews.csv')
```

```python
negative_dataset=dataset.query('label==0').reset_index(drop=True)
```

```python
negative_dataset
```

```python
corpus = []
for i in range(0, len(negative_dataset)):
    review = re.sub('[^a-zA-Z]', ' ', negative_dataset['review'][i])
    review = review.lower()
    review = review.split()
    ps = PorterStemmer()
    all_stopwords = stopwords.words('english')
    all_stopwords.remove('not')
    #review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
    review = [word for word in review if not word in set(all_stopwords)]
    review = ' '.join(review)
    corpus.append(review)
```

```python
text= " ".join(corpus) 
```

```python
r = Rake(min_length=1,max_length=1)
r.extract_keywords_from_text(text)

r.get_ranked_phrases()
```
